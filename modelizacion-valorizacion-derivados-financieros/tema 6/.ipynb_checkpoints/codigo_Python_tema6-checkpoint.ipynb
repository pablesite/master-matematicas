# Método de momentos estadísticos

```python
#Dt = 1
import numpy as np

S=np.array([29.56,30.125,28.56,27.50,28.125,27.44,
            27.06,27.125,28.69,27.94,27.25])

U=np.log(S[1:]/S[:-1])

sigmaMME=np.std(U,ddof=1) #ddof=1, 1/(N-ddof)*np.sum((U-np.mean(U))**2)
muMME=np.mean(U)+sigmaMME**2/2

# Mostrar estimaciones
print(f'muMME={muMME:0.5f} y sigmaMME={sigmaMME:0.5f}')
```

# Método de máxima verosimilitud

```python
#Dt = 1 
import numpy as np

S = np.array([29.56,30.125,28.56,27.50,28.125,27.44,
              27.06,27.125,28.69,27.94,27.25])
R = S[1:]/S[:-1]-1

muMMV = np.mean(R)
sigmaMMV = np.std(R,ddof=0)

# Mostrar estimaciones
print(f'muMMV={muMMV:0.5f} y sigmaMMV={sigmaMMV:0.5f}')
```

# Método de momentos no paramétrico

```python
# Dt = 1
import numpy as np

S = np.array([29.56,30.125,28.56,27.50,28.125,27.44,
              27.06,27.125,28.69,27.94,27.25])

R=S[1:]-S[:-1]

muMMNP=np.sum(R)/np.sum(S[:-1])
sigmaMMNP=np.sqrt(np.sum(R**2)/np.sum(S[:-1]**2))

print(f'muMMNP={muMMNP:0.5f} y sigmaMMNP={sigmaMMNP:0.5f}\n')
```

```python
# Mostrar todos

print(f'muMME={muMME:0.5f} y sigmaMME={sigmaMME:0.5f}\n')
print(f'muMMV={muMMV:0.5f} y sigmaMMV={sigmaMMV:0.5f}\n')
print(f'muMMNP={muMMNP:0.5f} y sigmaMMNP={sigmaMMNP:0.5f}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Cargar los datos desde el archivo CSV 'precios.csv'
df = pd.read_csv('precios.csv')

# Mostrar el DataFrame
print(df)

# Graficar los datos
plt.figure(figsize=(10, 5))
plt.plot(df['Fecha'], df['Precio'])
plt.title('Precios a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor legibilidad
plt.grid(True)
plt.tight_layout()  # Ajustar el diseño para evitar recortes
plt.show()
```

```python
# Calcular los rendimientos diarios
S = df['Precio'].values

R = S[1:] / S[:-1]

# Estimadores MME
U = np.log(R)

sigmaMME = np.std(U,ddof=1) #ddof=1, 1/(N-ddof)*np.sum((U-np.mean(U))**2)
muMME = np.mean(U)+sigmaMME**2/2

# Estimadores MMV
R -= 1
muMMV = np.mean(R)
sigmaMMV = np.std(R,ddof=0)

# Estimadores MMNP

R = S[1:] - S[:-1]

muMMNP = np.sum(R)/np.sum(S[:-1])
sigmaMMNP = np.sqrt(np.sum(R**2)/np.sum(S[:-1]**2))

# Mostrar todos

print(f'muMME={muMME:0.5f} y sigmaMME={sigmaMME:0.5f}\n')
print(f'muMMV={muMMV:0.5f} y sigmaMMV={sigmaMMV:0.5f}\n')
print(f'muMMNP={muMMNP:0.5f} y sigmaMMNP={sigmaMMNP:0.5f}')
```

# Modelo Log-Normal aleatorio (Wiener)

```python
#modelo Log-Normal
def St(t, mu, sigma, s0):
    n=len(t)
    
    # Wiener
    w = np.zeros(n)
    for i in range(1, n):
        w[i] = w[i - 1] + np.sqrt(t[i] - t[i-1])*np.random.normal(0,1)
    return s0*np.exp((mu-sigma**2/2)*t+sigma*w)
```

```python
n = len(S)
t = np.arange(0, n)

#ERROR MME
SSt = St(t, muMME, sigmaMME, S[0])

ECMMME = np.sqrt(np.mean((S[1:]-SSt[1:])**2))
EPAMMME = 100*np.mean(abs(1-SSt[1:]/S[1:]))

#ERROR MMV
SSt=St(t,muMMV,sigmaMMV,S[0])
ECMMMV=np.sqrt(np.mean((S[1:]-SSt[1:])**2))
EPAMMMV=100*np.mean(abs(1-SSt[1:]/S[1:]))

#ERROR MMNP
SSt = St(t, muMMNP, sigmaMMNP, S[0])
ECMMMNP = np.sqrt(np.mean((S[1:]-SSt[1:])**2))
EPAMMMNP = 100*np.mean(abs(1-SSt[1:]/S[1:]))

# Resultados
print(f'ECMMME={ECMMME:.5f} y EPAMMME(%)={EPAMMME:.3f}%')
print(f'ECMMMV={ECMMMV:.5f} y EPAMMMV(%)={EPAMMMV:.3f}%')
print(f'ECMMMNP={ECMMMNP:.5f} y EPAMMMNP(%)={EPAMMMNP:.3f}%')
```

# Media y varianza

```python
#mu_S(t) de Log-Normal
def ELN(t,mu,s0):
    return s0*np.exp(mu*t)

#sigma^2_S(t) de Log-Normal
def VLN(t,mu,sigma,s0):
    return s0**2*np.exp(2*mu*t)*(np.exp(sigma**2*t)-1)
```

# Intervalo de confianza al 95%

```python
#IC 95% MME
ELNMME = ELN(t, muMME, S[0])
SMME = np.sqrt(VLN(t, muMME, sigmaMME, S[0]))

# Crear el gráfico
fig, ax = plt.subplots()
ax.plot(t, ELNMME - 1.96 * SMME, '--r', label='IC inf')
ax.plot(t, ELNMME, 'g', label=r'$\mu_S(t)$')
ax.plot(t, ELNMME + 1.96 * SMME, 'r', label='IC sup')
ax.plot(t, S, 'b', label='S(t)')
ax.set_xlabel('t')
ax.set_title(r'IC 95% MME')
ax.legend()
plt.show()
```

```python
#IC 95% MMV
ELNMMV = ELN(t, muMMV, S[0])
SMMV = np.sqrt(VLN(t, muMMV, sigmaMMV, S[0]))

# Crear el gráfico
fig, ax = plt.subplots()
ax.plot(t, ELNMMV - 1.96 * SMMV, '--r', label='IC inf')
ax.plot(t, ELNMMV, 'g', label=r'$\mu_S(t)$')
ax.plot(t, ELNMMV + 1.96 * SMMV, 'r', label='IC sup')
ax.plot(t, S, 'b', label='S(t)')
ax.set_xlabel('t')
ax.set_title(r'IC 95% MMV')
ax.legend()
plt.show()
```

```python
#IC 95% MMV
ELNMMNP = ELN(t, muMMNP, S[0])
SMMNP = np.sqrt(VLN(t, muMMNP, sigmaMMNP, S[0]))

# Crear el gráfico
fig, ax = plt.subplots()
ax.plot(t, ELNMMNP - 1.96 * SMMNP, '--r', label='IC inf')
ax.plot(t, ELNMMNP, 'g', label=r'$\mu_S(t)$')
ax.plot(t, ELNMMNP + 1.96 * SMMNP, 'r', label='IC sup')
ax.plot(t, S, 'b', label='S(t)')
ax.set_xlabel('t')
ax.set_title(r'IC 95% MMNP')
ax.legend()
plt.show()
```

#Estimaciones

```python
m = len(S)//2

dt = S

S = S[:m]

R = S[1:] / S[:-1]

# Estimadores MME
U = np.log(R)

sigmaMME = np.std(U,ddof=1) #ddof=1, 1/(N-ddof)*np.sum((U-np.mean(U))**2)
muMME = np.mean(U)+sigmaMME**2/2

# Estimadores MMV
R -= 1
muMMV = np.mean(R)
sigmaMMV = np.std(R,ddof=0)

# Estimadores MMNP

R = S[1:] - S[:-1]

muMMNP = np.sum(R)/np.sum(S[:-1])
sigmaMMNP = np.sqrt(np.sum(R**2)/np.sum(S[:-1]**2))

# Mostrar todos

print(f'muMME={muMME:0.5f} y sigmaMME={sigmaMME:0.5f}\n')
print(f'muMMV={muMMV:0.5f} y sigmaMMV={sigmaMMV:0.5f}\n')
print(f'muMMNP={muMMNP:0.5f} y sigmaMMNP={sigmaMMNP:0.5f}')
```

```python
n = len(dt)
t = np.arange(0, n)

#Estimación MME
SMME = St(t, muMME, sigmaMME, S[0])

#Estimación MMV
SMMV=St(t,muMMV,sigmaMMV,S[0])

#Estimación MMNP
SMMNP = St(t, muMMNP, sigmaMMNP, S[0])

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(t, dt, 'b', label='S(t)', linewidth=2)
plt.plot(t, SMME, 'r', label='SMME', linewidth=2)
plt.plot(t, SMMV, 'g', label='SMMV', linewidth=2)
plt.plot(t, SMMNP, 'm', label='SMMNP', linewidth=2)
plt.xlabel('t')
plt.ylabel('Valor')
plt.title('Predicciones MME, MMV y MMNP')
plt.legend()
plt.grid(True)
plt.show()


# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(t, SMME - dt, 'r', label='SMME', linewidth=2)
plt.plot(t, SMMV - dt, 'g', label='SMMV', linewidth=2)
plt.plot(t, SMMNP - dt, 'm', label='SMMNP', linewidth=2)
plt.xlabel('t')
plt.ylabel('Valor')
plt.title('Predicciones MME, MMV y MMNP')
plt.legend()
plt.grid(True)
plt.show()

print(f'Errores de la estimación ||S-SMME||={np.linalg.norm(dt-SMME):.2e}, ||S-SMMV||={np.linalg.norm(dt-SMMV):.2e} y ||S-SMMNP||={np.linalg.norm(dt-SMMNP):.2e}')
```
