# Representación  de una  recta, una parábola  y una función cúbica

En este ejercicio se van a representar las funciones 
\begin{equation}y=x, 
y=x^2, 
y=x^3
\end{equation}
El objetivo de esta  representación es mostrar la forma más simple  de graficar las funciones.

# Una figura, un area única (axe) para las tres gráficas

Se importan las bibliotecas  necesarias y se define  el rango de valores de x

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

x = np.linspace(-6, 6)
```

1. Se crea la figura y  un único axes donnde se bvan a insertar las tres gráficas
2. Se crea cada una de las graficas y las etiquetas del eje de las x y de las y, además del título de la  figura.

```python
figura,graficas = plt.subplots()



graficas.plot(x, x, label='recta')  
graficas.plot(x, x**2, label='parabola')  
graficas.plot(x, x**3, label='cubica')  
#etiquetas
graficas.set_xlabel('x')  
graficas.set_ylabel('y')  
graficas.set_title("Gráfico sencillo") 
#añade la leyenda
graficas.legend()
plt.show()
```

# Una figura, tres areas (axes) una para cada gráfica

Se crea la figura y tres areas dentro (axes).
El rango  de las x es el mismo que se ha definido anteriormente.
Se ponen etiquetas en los ejes y título a cada una de las gráficas

```python
fig,ax=plt.subplots(1,3)
#una fila y tres columnas
fig.set_size_inches(10,8)
#se crea una recta en color rojo
ax[0].plot(x,x,label='recta', color="red" )
#se crea una parábola en color por defecto
ax[1].plot(x,x**2,label='parabola')
#se crea una función cúbica en color verde
ax[2].plot(x,x**3,label='cubica', color="green")

for i in  range(3):
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y')
    
ax[0].set_title("recta")
ax[1].set_title("parabola")
ax[2].set_title("cubica")
plt.show()
```
