# Ejercicio 5. Representación gráfica de tres funciones

En este ejercicio se van a representar sus funciones seno, coseno y producto de seno por exponencial mostrando su  amplitud.
El objetivo de esta  representación no es otro que el mostrar  el uso del graficador de Python y  los elementos que permiten  poner títulos y  nombres  además de leyendas. No se proporciona ninguna interpretación de  las funciones.

Primero se definen las tres  funciones

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    y = 8*np.sin(x)
    return y

def f2(x):
    y = 2*np.cos(x) 
    return y


def f3(x):
    y = np.cos(x)*np.exp(-x/10)
    return y
```

Se define el rango de valores que va a tomar x

```python
x = np.linspace(0, 10*np.pi, 100)
```

1. Se crean las gráficas y se crean las leyendas que las representan.
Se podría haber  hecho  también por separado
p1 = plot(x, f1(x), label='Funcion 1')
p2 = plot(x, f2(x), label='Funcion 2')
p3 = plot(x, f3(x), label='Funcion 3')

2. Se crean las etiquetas  de los ejes y el título
3. Se crea la ventana (figura) indicando el tamaño y se muestra


```python
p1, p2, p3 = plt.plot(x, f1(x), x, f2(x), x, f3(x))

# Leyendas especificando tamaño de letra 10, en esquina inferior derecha
plt.legend(('Funcion seno', 'Funcion coseno', 'Funcion exponencial'),
prop = {'size': 10}, loc='lower right')
#Se crean las etiquetas  de los ejes y el título
plt.xlabel('Tiempo en segundos')
plt.ylabel('Amplitud en cm')
plt.title('Representacion de tres funciones')
#Se crea la ventana (figura) indicando el tamaño y se muestra
figura=plt.figure(figsize=(200, 100))

plt.show()
```
