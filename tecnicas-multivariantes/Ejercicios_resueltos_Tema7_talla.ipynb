<h1>Talla (KNN)</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = {'altura': [158, 158, 158, 160, 160, 162, 163, 160, 163, 165, 165, 165, 168, 168, 167, 170, 170, 170],
        'peso': [58, 59, 63, 59, 61, 60, 61, 64, 64, 61, 62, 65, 62, 63, 66, 63, 64, 68],
        'talla': ['M', 'M', 'M', 'M', 'M', 'M', 'M', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L', 'L']}

df = pd.DataFrame(data)

punto_nuevo = {'altura': [160],
               'peso': [63]}

punto_nuevo = pd.DataFrame(punto_nuevo)

ax = plt.axes()
ax.scatter(df.loc[df['talla'] == 'L', 'altura'],
           df.loc[df['talla'] == 'L', 'peso'],
           c="green",
           label="L")
ax.scatter(df.loc[df['talla'] == 'M', 'altura'],
           df.loc[df['talla'] == 'M', 'peso'],
           c="blue",
           label="M")
ax.scatter(punto_nuevo['altura'],
           punto_nuevo['peso'],
           c="red")
for i, row in df.iterrows():
    ax.text(row['altura'], row['peso'], row['talla'], 
             ha='left', va='bottom')
plt.xlabel("Peso (kg)")
plt.ylabel("Altura (cm)")
ax.legend()
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=1) # ¿y si cambia n_neighbors?
X = df[['altura', 'peso']]
y = df[['talla']]
knn.fit(X, y)
prediccion = knn.predict(punto_nuevo)
print(prediccion)
```
