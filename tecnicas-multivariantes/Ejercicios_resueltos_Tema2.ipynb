```python
# EJERCICIO 2

# Apartado a)
# definir semilla para que los resultados sean los mismos--------
np.random.seed(seed = 11)
# se obtiene la muestra normal-----------------------------------
muestra_normal = np.random.normal(0, 1, 500)
muestra_unif = np.random.randint(0, 200, 500)
# para la muestra y1---------------------------------------------
y1 = 100 + 200 * muestra_unif + muestra_normal * muestra_unif
# Se necesita trasponer el vector x------------------------------
x = muestra_unif.reshape((-1, 1))
lm1 = linear_model.LinearRegression()
model1 = lm1.fit(x, y1)
# obtener valores modelo-----------------------------------------
y_pred1 = model1.predict(x)
# obtener residuos-----------------------------------------------
residuos1 = y1 - y_pred1
# para la muestra y2---------------------------------------------
y2 = 100 + 200 * muestra_unif + muestra_normal
lm2 = linear_model.LinearRegression()
model2 = lm2.fit(x, y2)
# obtener valores modelo-----------------------------------------
y_pred2 = model2.predict(x)
# obtener residuos-----------------------------------------------
residuos2 = y2 - y_pred2
plt.scatter(x, residuos1)
plt.axhline(0, color = "red", linewidth = 4)
plt.xlabel("x")
plt.ylabel("residuos")
plt.show()
plt.scatter(x, residuos2)
plt.axhline(0, color = "red", linewidth = 4)
plt.xlabel("x")
plt.ylabel("residuos")
plt.show()

# Apartado b)
# y1-------------------------------------------------------------
m1 = sm.OLS(y1, sm.add_constant(x)).fit()
bp1 = sms.het_breuschpagan(resid = m1.resid, 
exog_het = m1.model.exog)[1]
print("El resultado del test Breusch-Pagan es: p.valor = %5.3f" 
%(bp1))

# y2-------------------------------------------------------------
m2 = sm.OLS(y2, sm.add_constant(x)).fit()
bp2 = sms.het_breuschpagan(resid = m2.resid, 
exog_het = m2.model.exog)[1]
print("El resultado del test Breusch-Pagan es: p.valor = %5.3f" 
%(bp2))
```

Ejemplo de cálculo de estimadores con librerías

```python
import numpy as np
from scipy import stats

# Datos
X = np.array([4, 3, 11, 5, 4, 50, 6, 8, 7, 9])

# Media aritmética
mean_arithmetic = np.mean(X)

# Mediana
median_value = np.median(X)

# Moda
mode_value = stats.mode(X, keepdims=True).mode[0]

# Media recortada al 10%
trimmed_mean_10 = stats.trim_mean(X, 0.1)

# Media windsorizada al 10%
windsorized_X = stats.mstats.winsorize(X, limits=0.1)
windsorized_mean = np.mean(windsorized_X)

# Trimedia
Q1 = np.percentile(X, 25)
Q2 = np.percentile(X, 50)
Q3 = np.percentile(X, 75)
trimedia = (Q1 + 2 * Q2 + Q3) / 4

# Media geométrica
geo_mean = stats.gmean(X)

# Media armónica
harm_mean = stats.hmean(X)

# Media de Huber (parametro de tuning default)
huber_mean = stats.tmean(X, (np.percentile(X, 10), np.percentile(X, 90)))

# Cálculo de la Centrimedia
P25 = np.percentile(X, 25)
P37_5 = np.percentile(X, 37.5)
P50 = np.percentile(X, 50)
P62_5 = np.percentile(X, 62.5)
P75 = np.percentile(X, 75)

centrimedia = (P25 + P37_5 + P50 + P62_5 + P75) / 5


# Resultados
estimadores = {
    "Media Aritmética": mean_arithmetic,
    "Mediana": median_value,
    "Moda": mode_value,
    "Media Recortada (10%)": trimmed_mean_10,
    "Media Windsorizada (10%)": windsorized_mean,
    "Trimedia": trimedia,
    "Media Geométrica": geo_mean,
    "Media Armónica": harm_mean,
    "Media de Huber": huber_mean,
    "Centrimedia": centrimedia
}

import pandas as pd
df_estimadores = pd.DataFrame(list(estimadores.items()), columns=["Estimador", "Valor"])
import ace_tools as tools
tools.display_dataframe_to_user(name="Estimadores Estadísticos", dataframe=df_estimadores)
```
